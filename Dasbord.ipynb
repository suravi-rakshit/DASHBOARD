```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("dasbord.xlsx")
df=pd.DataFrame(data)
print(data)
plt.pie(df["New users"],labels=df["Duration Day 1"])
plt.xlabel("New users",fontsize=20)
plt.ylabel("Duration Day 1",fontsize=20)
plt.title("Cohort Analysis",fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("dasbord.xlsx")
df=pd.DataFrame(data)
print(data)
grouped_by=df.groupby("New users")["Duration Day 1"].sum()
print(grouped_by)
colors=["red","blue","green"]
plt.bar(grouped_by.index,grouped_by.values,color=colors)
plt.xlabel("New users",fontsize=20)
plt.ylabel("Duration Day 1",fontsize=20)
plt.title("Cohort Analysis",fontsize=20)
plt.show()
```
